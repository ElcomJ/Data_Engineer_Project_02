```python
import boto3
import json
from random import uniform
import time
from datetime import datetime

client = boto3.client('kinesis', \
                      aws_access_key_id     = 'YOUR_ACCESS_KEY_ID_HERE', \
                      aws_secret_access_key = 'YOUR_ACCESS_SECRET_KEY_HERE', \
                      region_name           = 'YOUR_SERVICE_REGION_NAME_HERE')
```

```python
id = 0
while True:
  data = uniform(70, 80)
  id += 1;
  reg = {'idtemp' : str(id), \
         'data' : str(data), \
         'type' : 'hydraulic_pressure', \
         'timestamp' : str(datetime.now())}        
  client.put_record(StreamName   = 'YOUR_DATA_STREAM_NAME_HERE', \
                    Data         = json.dumps(reg), \
                    PartitionKey = '02')
  
  time.sleep(10)
  print(reg)
```
